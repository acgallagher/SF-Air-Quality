```python
import logging
import boto3
from botocore.exceptions import ClientError
import os
from dotenv import dotenv_values

def upload_file(file_name, bucket, object_name=None):
    """Upload a file to an S3 bucket

    :param file_name: File to upload
    :param bucket: Bucket to upload to
    :param object_name: S3 object name. If not specified then file_name is used
    :return: True if file was uploaded, else False
    """

    # If S3 object_name was not specified, use file_name
    if object_name is None:
        object_name = os.path.basename(file_name)

    # Load AWS Credentials
    config = dotenv_values("aws_credentials.env")
    print(config)
    s3_client = boto3.client(
        's3',
        aws_access_key_id=config["access_key"],
        aws_secret_access_key=config["secret_access_key"]
        )
    
    # Upload the file
    try:
        response = s3_client.upload_file(file_name, bucket, object_name)
    except ClientError as e:
        logging.error(e)
        return False
    return True

upload_file("data/sensors.parquet", "sf-air-quality-bucket")
```

```python
from datetime import datetime
import json
import requests

def extract_sensor_data(sensor_id: str) -> dict:
    start = datetime(2020, 9, 10).timestamp()
    end = datetime(2020, 9, 11).timestamp()
    fields = "humidity, temperature, pressure, \
        pm1.0_atm, pm2.5_atm, pm10.0_atm,  \
        scattering_coefficient, deciviews, visual_range"

    url = f"https://api.purpleair.com/v1/sensors/{sensor_id}/history"
    api_key = "7CD6D264-0A25-11EE-BD21-42010A800008"
    headers = {"X-API-Key":api_key}
    payload = {
        "start_timestamp":start,
        "end_timestamp":end, 
        "average":30,
        "fields":fields,        
            }

    r = requests.get(url, headers=headers, params=payload)
    data = r.json()

    return data

orange_day_data = []
count = 0
with open("data/sensors copy.csv", "r") as f:
    f.readline()
    for line in f.readlines():
        print(count)

        sensor_id = line.split(",")[0]
        historical_data = extract_sensor_data(sensor_id)
        orange_day_data.append(historical_data)

        count += 1


print(orange_day_data)

with open("data/orange_day_data_testing_10_16_2.json", "w") as f:
    json.dump(orange_day_data, f, ensure_ascii=False, indent=2)
        
```

```python
import json
import pandas as pd
from datetime import timedelta

with open("data/orange_day_data_testing_10_16_2.json", "r") as f:
    historical_data = json.load(f)

historical_data_list = []
historical_data_fields = ["sensor_index","time_stamp", "humidity", "temperature", "pressure", "pm1.0_atm", "pm2.5_atm", "pm10.0_atm", "scattering_coefficient", "deciviews", "visual_range"]
for sensor_data in historical_data:
    if "error" in sensor_data:
        continue
    elif len(sensor_data["data"]) == 0: # skip if there is no data
        continue

    sensor_index = sensor_data["sensor_index"]

    for time_slice in sensor_data["data"]:
        time_slice.insert(0, sensor_index)
        historical_data_list.append(time_slice)

df = pd.DataFrame(historical_data_list, columns=historical_data_fields)
df.time_stamp = df.time_stamp.apply(
        lambda x: datetime(1970, 1, 1) + timedelta(seconds=x) - timedelta(hours=4)
    )
df.to_csv("data/orange_day_data_testing_10_16_2.csv", index=False)
```
